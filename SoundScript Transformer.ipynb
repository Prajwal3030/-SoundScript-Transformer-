```python
from gtts import gTTS
from pydub import AudioSegment
from pydub.playback import play
import time
import speech_recognition as sr
from google.colab import files


r = sr.Recognizer()

def record_text():
    print("Please enter your text:")
    return input()

def record_audio_from_upload():
    print("Upload an MP3 file:")
    uploaded_file = files.upload()

    if len(uploaded_file) == 0:
        print("No file uploaded. Please try again.")
        return None

    file_path = list(uploaded_file.keys())[0]
    return AudioSegment.from_file(file_path)

def convert_audio_to_text(audio):
    temp_wav_path = "temp.wav"
    audio.export(temp_wav_path, format="wav")

    with sr.AudioFile(temp_wav_path) as source:
        r.adjust_for_ambient_noise(source)
        audio_data = r.record(source)
        return r.recognize_google(audio_data)

def output_text_to_file(text, file_path):
    with open(file_path, "a") as f:
        f.write(text + "\n")

def convert_text_to_audio(text, output_path='output.mp3'):
    tts = gTTS(text=text, lang='en', slow=False)
    tts.save(output_path)

    try:
        audio = AudioSegment.from_mp3(output_path)
        play(audio)
        print("Audio file playback initiated.")
    except Exception as e:
        print(f"Error playing audio file: {e}")

output_file_path = "output.txt"
try:
    while True:
        print("Choose an option:")
        print("1. Enter text")
        print("2. Upload an MP3 file")
        print("3. Exit")
        option = input().strip()

        if option == "1":

            text = record_text()
        elif option == "2":

            audio = record_audio_from_upload()
            if audio is None:
                continue
            text = convert_audio_to_text(audio)
        elif option == "3":
            print("Exiting the code.")
            break
        else:
            print("Invalid option. Please choose 1, 2, or 3.")
            continue


        if text:
            output_text_to_file(text, output_file_path)
            print(f"Converted text: {text}")


            convert_text_to_audio(text)


            time.sleep(2)

except KeyboardInterrupt:
    print("Execution manually interrupted.")
```
